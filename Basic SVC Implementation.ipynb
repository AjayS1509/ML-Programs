```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##lets create as synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
x1 = pd.DataFrame(X)[0]
x2 = pd.DataFrame(X)[1]
```

```python
sns.scatterplot(x =x1,y =x2, hue=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc = SVC(kernel="linear")
```

```python
svc.fit(X_train, y_train)
```

```python
##prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf =SVC(kernel="rbf")
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred1 = rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
poly =SVC(kernel="poly")
poly.fit(X_train,y_train)
y_pred2 = poly.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sig =SVC(kernel="sigmoid")
sig.fit(X_train,y_train)
y_pred3 = poly.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
##Hyper parameter tunbning with SVC
from sklearn.model_selection import GridSearchCV

#defining parameter range
param_grid = {'C': [0.1,1,10,100,1000],
                'gamma':[1,0.1,0.01,0.001,0.0001],
                 'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
#predictions
y_pred4 = grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
