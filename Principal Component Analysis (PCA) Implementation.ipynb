```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
##Load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
cancer_dataset.DESCR
```

```python
df = pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset.feature_names)
```

```python
df.head()
```

```python
##Standaradization
from sklearn.preprocessing import StandardScaler
sclaer = StandardScaler()
```

```python
sclaer.fit(df)
```

```python
scalar_data = sclaer.transform(df)
```

```python
scalar_data
```

```python
##Apllying pca algorithm
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(scalar_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1], c = cancer_dataset.target, cmap='plasma')
plt.xlabel('First pricipal component')
plt.ylabel('Second principal component')
plt.show()
```
