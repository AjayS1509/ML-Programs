```python
from sklearn.datasets import load_diabetes
```

```python
datasets = load_diabetes()
```

```python
datasets
```

```python
print(datasets['DESCR'])
```

```python
import pandas as pd
```

```python
df_diabetes = pd.DataFrame(datasets.data, columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
```

```python
X = df_diabetes
```

```python
X
```

```python
y = datasets.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
##Hyper parameter Tunning
params_grid = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
grid = GridSearchCV(regressor, param_grid=params_grid,cv = 5, scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
selectedmodel = DecisionTreeRegressor(criterion='friedman_mse',max_depth=4, max_features='log2',splitter='best')
selectedmodel.fit(X_train, y_train)
```

```python
##Visualize the decison tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedmodel, filled=True)
plt.show()
```
